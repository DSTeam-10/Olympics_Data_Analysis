## Data Cleaning

```python
# List the columns
df_athlete.columns
```

```python
irrelevant_cols=["ID", "Name", "Height", "Weight", "Season", "City"]
print(f'Before Removing Irrelevant Columns : {df_athlete.shape}')
df_athlete = df_athlete.drop(irrelevant_cols, axis=1)
print(f'After Removing Irrelevant Columns : {df_athlete.shape}')
df_athlete.head()
```

## One hot encoder

<p>Creating a one-hot encoded representation for each distinct medal category. </p>

```python
from sklearn.preprocessing import OneHotEncoder
df_athlete['Medal'].fillna('No Medal', inplace=True)

encoder = OneHotEncoder(sparse=False, drop=None, categories='auto')  # 'drop' parameter is set to ignore the first category
medal_encoded = encoder.fit_transform(df_athlete[['Medal']])
medal_categories = encoder.get_feature_names_out(['Medal'])

df_athlete[medal_categories] = medal_encoded

df_athlete.head()
```

```python
# Get the total sum of each medal type
team_medals = df_athlete.groupby('Team').agg({
    'Medal_Gold': 'sum',
    'Medal_Silver': 'sum',
    'Medal_Bronze': 'sum'
}).reset_index()

df_athlete = pd.merge(df_athlete, team_medals, on='Team', how='left')

df_athlete.rename(columns={
    'Medal_Gold_y': 'Total_Gold_Medals',
    'Medal_Silver_y': 'Total_Silver_Medals',
    'Medal_Bronze_y': 'Total_Bronze_Medals'
}, inplace=True)

df_athlete.drop(['Medal_Gold_x', 'Medal_Silver_x', 'Medal_Bronze_x'], axis=1, inplace=True)
```

```python
# creating total medals won by summing up each distinct medal category
df_athlete['Total_Historical_Medals'] = df_athlete['Total_Gold_Medals'] + df_athlete['Total_Silver_Medals'] + df_athlete['Total_Bronze_Medals']
```

<p>The 'Total_Medal_Value' column in the DataFrame is calculated as the weighted sum of gold, silver, and bronze medals, assigning higher values to gold (3), silver (2), and bronze (1). This creates a composite measure reflecting the overall success of each team based on their medal achievements.</p>

```python

df_athlete['Total_Medal_Value'] =  df_athlete['Total_Gold_Medals']*3 + df_athlete['Total_Silver_Medals']*2 + df_athlete['Total_Bronze_Medals']
```

```python
# Plot correlation heatmap
sns.heatmap(data=df_athlete.corr(numeric_only=True), annot=True)
```

<p>The dataset is consolidated by retaining the first entry for each unique team, providing a comprehensive overview of total medal performance for distinct teams in the updated DataFrame.</p>

```python
# keep distinct team values
df_athlete = df_athlete.groupby('Team').first().reset_index()
df_athlete.head()
```

## Build Model

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select the feature for clustering (Total_Medal_Value)
X = df_athlete[['Total_Medal_Value']]

# Standardize the feature for better clustering results
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Determining Total Clusters using Elbow Method

```python
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.show()
```

### Apply K-means Clustering Model

```python
# Choose the number of clusters
num_clusters = 3 

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df_athlete['Cluster'] = kmeans.fit_predict(X_scaled)

# Display the clusters and corresponding sponsorship predictions
cluster_means = df_athlete.groupby('Cluster')['Total_Medal_Value'].mean().sort_values()

cluster_labels_mapping = {0: 'Low', 1: 'Medium', 2: 'High'}

df_athlete['Sponsorship_Prediction'] = df_athlete['Cluster'].map(cluster_labels_mapping)

# Display the final DataFrame with predicted sponsorship
df_athlete[['Team', 'Total_Medal_Value', 'Cluster', 'Sponsorship_Prediction']]
```

<h4>Visualize the clusters in scatter plot</h4>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(10, 6))

for cluster_label in df_athlete['Cluster'].unique():
    cluster_data = df_athlete[df_athlete['Cluster'] == cluster_label]
    sns.scatterplot(x='Total_Medal_Value', y='Cluster', data=cluster_data, label=f'Cluster {cluster_label}', alpha=0.8)

plt.scatter(kmeans.cluster_centers_[:, 0], range(num_clusters),s=100, color='red', label='Cluster Centers')

plt.xlabel('Total_Medal_Value')
plt.ylabel('Cluster')
plt.show()
```

### Export Model
